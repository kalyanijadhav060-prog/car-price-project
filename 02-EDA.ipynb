```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#import dataset
df=pd.read_csv("cleaned_data.csv")
df.info()
```

```python
#company-wise count of vehicles
data=df["company"].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data.index,data.values,color="blue")
plt.xticks(rotation=90)
plt.xlabel("companies")
plt.ylabel("count of vehicles")
plt.title("Company-wise count of vehicles")
for X,y in zip(data.index,data.values):
    plt.text(X,y,y,ha="center",va="bottom",fontsize=6)
plt.show()
```

```python
#yearwise count of vehicles
data=df["year"].value_counts().sort_index()
ax=data.plot(kind="bar",figsize=(12,5),color="cyan")
ax.bar_label(ax.containers[0],fontsize=6,fontweight="bold")
plt.xlabel("year")
plt.ylabel("count of vehicles")
plt.title("year-wise count of vehicles")
plt.show()
```

```python
#we will keep vehicel runnig between 1000 and 200000
df=df[(df["kms_driven"]>=1000)&(df["kms_driven"]<=200000)]
df.info()
```

```python
#kms_driven plot
sns.swarmplot(x="company",y="kms_driven",data=df)
plt.xticks(rotation=90)
plt.show()
df.info()
```
